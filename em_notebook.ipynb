```python
def convert_to_float(value):
    # Remove periods and replace commas with periods
    cleaned_value = value.replace('.', '').replace(',', '.')
    return float(cleaned_value)


data['Marktwert'] = data['Marktwert'].apply(convert_to_float)


def Strength(team):
    filtered_data = data[data['Land'] == team].copy()
    #filtered_data['GoalValue'] = filtered_data['GoalValue'].str.replace(',', '.')
    filtered_data['Goals'] = filtered_data['Goals'].astype(int)
    strength_value_sum = filtered_data['Goals'].sum()
    return strength_value_sum
    

def Chance(team):
    filtered_data = data[data['Land'] == team].copy()
    filtered_data['GCA'] = filtered_data['GCA'].astype(int)
    chance_value_sum = filtered_data['GCA'].sum()
    return chance_value_sum


def Fairness(team):
    counter = 0
    filtered_data = data[data['Land'] == team].copy()
    filtered_data['Gelb'] = filtered_data['Gelb'].astype(int)
    filtered_data['Rot'] = filtered_data['Rot'].astype(int)
    for value in filtered_data['Rot']:
        counter += value*5
    fairness_value_sum = filtered_data['Gelb'].sum()
    fairness_value_sum += counter
    return fairness_value_sum


def Experience(team):
    filtered_data = data[data['Land'] == team].copy()
    filtered_data['TimeSum'] = filtered_data['TimeSum'].astype(int)
    exp_value_sum = filtered_data['TimeSum'].sum()
    return exp_value_sum


def Alter(team):
    filtered_data = data[data['Land'] == team].copy()
    filtered_data['Age'] = filtered_data['Age'].astype(int)
    age_value_sum = filtered_data['Age'].sum()
    return age_value_sum


def League(team, mult: int):
    counter = 0
    filtered_data = data[data['Land'] == team].copy()
    for league in filtered_data['Liga']:
        # Check if the league is one of the desired leagues
        if league == "Serie A":
            counter += 1
        elif league == "Bundesliga":
            counter += 1
        elif league == "Premiere League" or league == "Premiere League ":
            counter += 1
        elif league == "La Liga":
            counter += 1
        elif league == "Ligue 1":
            counter += 1
    return counter

def Value(team):
    filtered_data = data[data['Land'] == team].copy()
    filtered_data['Marktwert'] = filtered_data['Marktwert'].astype(float)
    market_value_sum = filtered_data['Marktwert'].sum()
    return market_value_sum


def Taktics(team):
    counter = 0
    filtered_data = data[data['Land'] == team].copy()
    for position in filtered_data['Position']:
        # Check if the league is one of the desired leagues
        if position == "Verteidiger":
            counter += 0
        elif position == "Mittelfeld":
            counter += 0.5
        elif position == "Sturm":
            counter += 1
    return counter

def Defense(team):
    filtered_data = data[data['Land'] == team].copy()
    filtered_data['Save%'] = filtered_data['Save%'].str.replace(',', '.')
    filtered_data['Save%'] = filtered_data['Save%'].astype(float)
    chance_value_sum = filtered_data['Save%'].sum()
    return chance_value_sum

def DailyForm(num):
        if num >= 0 and num < 0.2:
            return -0.05
        elif num >= 0.2 and num < 0.4:
            return -0.1
        elif num >= 0.4 and num < 0.6:
            return 0.05
        elif num >= 0.6 and num < 0.8:
            return 0.1
        else:
            return 0


Strength("Deutschland")
Chance("Deutschland")
Fairness("Deutschland")
Experience("Deutschland")
Alter("Deutschland")
League("Deutschland", 5)
Value("Deutschland")
Taktics("Deutschland")
Defense("Deutschland")
```
